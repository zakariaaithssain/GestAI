## OpenAI

Instead of `curl`, we can call the vLLM-hosted model through the OpenAI client SDK

```python
from openai import OpenAI #This not installed yet
```

```python
openai_api_key = "EMPTY" #We don't need an oepnai api key when querying a local model
openai_api_base = "http://localhost:8000/v1" #This the host:port we used to serve with vLLM
```

```python
#Create the server client
client = OpenAI(
        api_key=openai_api_key,
        base_url=openai_api_base,
    )
```

```python
messages = [
    {
        "role": "system", 
        "content": "You are a helpful assistant."},
    {
        "role": "user",
        "content": "What's the capital of France ?"
    }
]
```

```python
chat_completion = client.chat.completions.create(
    messages=messages,
    model="Qwen/Qwen3-0.6B",
)
```

```python
print(chat_completion)
```
